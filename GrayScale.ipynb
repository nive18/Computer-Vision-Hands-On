```python

import numpy as np
import matplotlib.image as mpimg  # for reading in images

import matplotlib.pyplot as plt
import cv2  # computer vision library

%matplotlib inline
```

```python

image = mpimg.imread('sunrise.jpg')
```

```python
print(image.shape)
```

```python
#Change from color to greyscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image,cmap='gray')
```

```python
print(gray_image[300,400])
```

```python
#If we want to use the maximum value accross a axi for example ([1,2,3],[2,3,4]) it will print ([3,4])
max_value = np.amax(gray_image)
min_value = np.amin(gray_image)
print(max_value,min_value)
```
