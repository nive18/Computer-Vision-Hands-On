```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
image = mpimg.imread('highway.jpg')
```

```python
plt.imshow(image)
```

```python
#plt.subplots() is a function that returns a tuple containing a figure and axes object(s). Thus when using fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax.
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25,40))
ax1.set_title('R channel')
ax1.imshow(r, cmap='Reds')
ax2.set_title('G channel')
ax2.imshow(g, cmap='Greens')
ax3.set_title('B channel')
ax3.imshow(b, cmap='Blues')
```
